```python
import pandas as pd
import numpy as np
import datetime
import calendar
from math import pi
from datetime import datetime

# Bokeh libraries

from bokeh.io import output_file, output_notebook
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource, LinearColorMapper, \
    BasicTicker, PrintfTickFormatter, ColorBar, HoverTool, \
    DatetimeTickFormatter, LinearAxis
from bokeh.layouts import row, column, gridplot
from bokeh.models.widgets import Tabs, Panel
```

```python
career_progress_file = 'data.csv'
dashboard_file = 'index.html'
```

```python
timeseries_df = pd.read_csv(career_progress_file)
timeseries_df['Root_tot'] = timeseries_df.Root.cumsum()
print(timeseries_df)
```

```python
line_cds = ColumnDataSource(timeseries_df[['Matches', 'Root','Root_tot']])

select_tools = ['box_select', 'box_zoom', 'lasso_select', 'tap', 'reset']

fig_lday = figure(
    title='Career Runs',
    x_axis_label='Matches',
    y_axis_label='Runs',
    x_minor_ticks=2,
    y_range=(0, timeseries_df.Root_tot.max() + 1000),
    tools=select_tools,
    toolbar_location='right',
    plot_width=1600,
    plot_height=800
    )

fig_lday.line(
    x='Matches',
    y='Root_tot',
    source=line_cds,
    color='blue',
    line_width=3,
    legend='Joe Root',
    )

fig_lday.circle(
    x='Matches',
    y='Root_tot',
    source=line_cds,
    color='black',
    selection_color='darkgreen',
    nonselection_color='lightgray',
    size=8,
    alpha=0.5,
    )

hover_glyph = fig_lday.circle(
    x='Matches',
    y='Root_tot',
    source=line_cds,
    hover_fill_color='black',
    hover_alpha=0.5,
    size=15,
    alpha=0,
    )

tooltips = [('Match Runs', '@Root'), ('Cumulative Runs', '@Root_tot'), ('Matche', '@Matches')]

fig_lday.add_tools(HoverTool(tooltips=tooltips,
                   renderers=[hover_glyph]))

output_file(dashboard_file)
show(fig_lday)
```
